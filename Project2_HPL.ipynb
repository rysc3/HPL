### Hopper 
We notice large variations on hopper for the NB values. This could be caused by two different things, firstly this can be coming via the network architecture. We spoke briefly about the issues being caused with Hopper's setup to where there is one star network node that branches off onto the other networking nodes, and then to the clusters. This adds significant latency, and could be the reason why my data resulted this way. The other thing that is important to keep in mind is because of other errors that came up, I was unable to get my runs to work with the parameters provided, so it could be that the runs which did particularly poorly just had a very less-than-optimal node spread and had to do a lot of communicating that way. Because we have a weaker networking architecutre that completely coinsides with our finding on the heat map; using a flatter grid, i.e. 1x8 (pxq) will provide better results because it will involve less communication compared to a more squared (i.3. 4x4) configuration. 

### Wheeler
Wheeler looks much closer to what my hypothesis was. Optimizing for the NB value I found an optimal result with 160. We see the overall speed decrease over time as I increase my problem size, this is likely because of sub-par network communication speeds. Then plugging in the optimized nb value to perform a parameter sweep for P & Q we see what is not a clear pattern for optimal values. My analysis on this is that because of how many jobs were running on wheeler I know for a fact that my jobs were multi-node runs on heavily fractured node configurations. This could definitely be the issue for the inconsistent data, but generally wheeler appears to have better results in teh center of the heat, or in other words, more squared. We see this verified throught the generated HPL.dat for wheeler that gave us an optimal value of 4x4.
